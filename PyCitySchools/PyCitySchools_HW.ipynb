```python
import pandas as pd
import numpy as np
#accessing file
load_school_data = "Resources/schools_complete.csv"
load_student_data = "Resources/students_complete.csv"

#assigning read file to variable
school_data = pd.read_csv(load_school_data)
student_data = pd.read_csv(load_student_data)

#merging student and school data into a table
school_data_table = pd.merge(student_data, school_data)
school_data_table
```

```python
#count the number of unique schools based on school ID
number_of_schools = len(school_data_table["School ID"].unique())
number_of_schools
```

```python
#count the number of unique students based on student ID
number_of_students = len(school_data_table["Student ID"].unique())
number_of_students
```

```python
#sum the budgets of all schools
total_budget = school_data["budget"].sum()
total_budget
```

```python
#average of all math scores
math_score_avg = student_data["math_score"].mean()
math_score_avg
```

```python
#average of all reading scores
reading_score_avg = student_data["reading_score"].mean()
reading_score_avg
```

```python
#average overall
avg_score_overall =(math_score_avg+reading_score_avg)/2
avg_score_overall
```

```python
#calculate percentage of students with a scoreover 70
#identify passing score
student_data["#passing_math"] = student_data["math_score"] >= 70
student_data["#passing_reading"] = student_data["reading_score"] >= 70
passing_math_percentage = ((student_data["#passing_math"]).mean())*100
passing_math_percentage
```

```python
passing_read_percentage = ((student_data["#passing_reading"]).mean())*100
passing_read_percentage 
```

```python
overall_passing_rate = (passing_math_percentage + passing_read_percentage)/2
overall_passing_rate
```

```python
#dataframe to hold all results
#assign the variable for each column
school_statistics = [{"Total Schools": number_of_schools, 
            "Total Students": number_of_students, 
            "Total Budget": total_budget, 
            "Average Math Score":  round(math_score_avg,2), 
            "Average Reading Score":  round(reading_score_avg,2), 
           "% Passing Math": round(passing_math_percentage,2),
           "% Passing Reading": round(passing_read_percentage ,2),
            "% Overall Passing": round(overall_passing_rate,2)}]
#assign the variables in a table
stat_summary_table = pd.DataFrame(school_statistics)

#
stat_summary_table["% Passing Math"] = stat_summary_table["% Passing Math"].map("{:,.2f}%".format)
stat_summary_table["% Passing Reading"] = stat_summary_table["% Passing Reading"].map("{:,.2f}%".format)
stat_summary_table["% Overall Passing"] = stat_summary_table["% Overall Passing"].map("{:,.2f}%".format)
stat_summary_table["Total Budget"] = stat_summary_table["Total Budget"].map("${:,.2f}".format)
stat_summary_table["Total Students"] = stat_summary_table["Total Students"].map("{:,}".format)

stat_summary_table
```

```python
#display complete table
school_data_table["passing_math"] = school_data_table["math_score"] >= 70
school_data_table["passing_reading"] = school_data_table["reading_score"] >= 70
school_data_table
```

```python
#group by school
school_group = school_data_table.groupby(["school_name"]).mean()

#Calculate new beneficial information
#Budget per student
school_group["Budget Per Student"] = school_group["budget"]/school_group["size"]
#Math pass rate per school
school_group["% Passing Math"] = round(school_group["passing_math"]*100,2)
#Read pass rate per school
school_group["% Passing Reading"] = round(school_group["passing_reading"]*100,2)
#Overall passing rate per school
school_group["% Overall Passing"] = round(((school_group["passing_math"] + school_group["passing_reading"])/2)*100,3)

#merge new data with existing data
school_stats = pd.merge(school_group, school_data, how="left", on=["school_name", "school_name"])

#clear specific variables
del school_stats['size_y']
del school_stats['budget_y']
del school_stats['Student ID']
del school_stats['School ID_x']

#new table
school_stats_table = pd.DataFrame({"School Name":  school_stats["school_name"],
                                "School Type": school_stats["type"],
                               "Number of Students": school_stats["size_x"],
                               "School Budget": school_stats["budget_x"],
                               "Budget Per Student": school_stats["Budget Per Student"], 
                               "Avg Math Score":round(school_stats["math_score"],2),
                               "Avg Reading Score":round(school_stats["reading_score"],2), 
                               "% Passing Math": school_stats["% Passing Math"],
                               "% Passing Reading": school_stats["% Passing Reading"],
                               "% Overall Passing": school_stats["% Overall Passing"]}) 

#Formatting
school_stats_table["Number of Students"] = school_stats_table["Number of Students"].map("{:,.0f}".format)
school_stats_table["School Budget"] = school_stats_table["School Budget"].map("${:,.2f}".format)
school_stats_table["Budget Per Student"] = school_stats_table["Budget Per Student"].map("${:,.2f}".format)

#Display
school_stats_table 
```

```python
#Organize top 5 schools and sort by % overall passing
top_five = school_stats_table.sort_values(["% Overall Passing"], ascending=False)
top_five.head()
```

```python
#Organize bottom 5 schools and sort by % overall passing
bottom_five = school_stats_table.sort_values(["% Overall Passing"], ascending=True)
bottom_five.head()
```

```python
#spending per student ranged from 578 to 652
spending_bins = [0, 580, 610, 640, 670]
bin_names = ["<$580", "$581-610", "$611-640", "$641-670"]

#name new column of Budget Per Student categorized by bins
school_stats["Spending Ranges (Per Student)"] = pd.cut(school_stats["Budget Per Student"], spending_bins, labels=bin_names)
grouped_school_spending = school_stats.groupby("Spending Ranges (Per Student)").mean() 

del grouped_school_spending['size_x']
del grouped_school_spending['budget_x']
del grouped_school_spending['Budget Per Student']
del grouped_school_spending['School ID_y']
del grouped_school_spending['passing_math']
del grouped_school_spending['passing_reading']

grouped_school_spending
```

```python
#performance categorized by school size
size_bins = [0, 1000, 2000, 5000]
bin_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#focusing on the column 'size'
school_stats["School Size"] = pd.cut(school_stats["size_x"], size_bins, labels=bin_names)

grouped_school_size = school_stats.groupby("School Size").mean() 

del grouped_school_size['budget_x']
del grouped_school_size['Budget Per Student']
del grouped_school_size['School ID_y']
del grouped_school_size['passing_math']
del grouped_school_size['passing_reading']

grouped_school_size
```
